```python
import pygame
from medialib import *


pygame.init()
initialize()  # always the first instruction of the program
##################################################################################

##  each item of the list representing the background of each row
background=[['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X'],
            ['X','_','_','L','_','_','M','_','_','E','_','_','J','_','_','_','_','X'],
            ['X','_','_','_','M','_','B','_','L','_','_','_','_','B','_','J','_','X'],
            ['X','_','_','B','_','_','_','_','_','J','_','_','E','_','L','_','_','X'],
            ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X']] 




def draw_cha(cha_row,x_pos,y_pos):
    cell=cha_row
    if cell=='X':
        draw('MazeGame_imgs/X.png',x_pos,y_pos)
    elif cell=='_':
        draw('MazeGame_imgs/_.png',x_pos,y_pos)
    elif cell=='L':
        draw('MazeGame_imgs/L.png',x_pos,y_pos)
    elif cell=='M':
        draw('MazeGame_imgs/M.png',x_pos,y_pos)
    elif cell=='B':
        draw('MazeGame_imgs/B.png',x_pos,y_pos)
    elif cell=='E':
        draw('MazeGame_imgs/E.png',x_pos,y_pos)
    elif cell=='J':
        draw('MazeGame_imgs/J.png',x_pos,y_pos)
        

        
collidables = 'X' ## the letters that the avatar cannot pass
monsters='M'## the letters that the avatar will dead when meeting it


x0=10
y0=200 
width=30
Score=0



welcome_msg = {'L': 'Lufthansa', 'E': 'EasyJet', 'J': 'Jet2'}

def draw_background():
    global x0, y0, Score 
    clear(200,200,200)  ## background
    for row in background:
        for cell in row:
            draw_cha(cell,x0,y0)
            x0+=width
        y0+=width
        x0=10
    y0=200
        
    text('Maze game',300,100,20)
    text('You are a lost hero, and have to find your way back home...',10,500,16)
    text('When you collect 20 points, you can get back to home!',10,520,16)
    text('To play: press a and d keys to move left and right',10,540,16)
    text('         press w and s keys to move upwards and downwards',10,560,16)
    text('         press q to quit the game',10,580,16)
    



def human(x,y):
    draw('MazeGame_imgs/H.png',x,y)
    global x0, y0, Score
    
    if background[(yh - y0) // width][(xh - x0) // width] in ['L', 'E', 'J']:
        airline_key = background[(yh - y0) // width][(xh - x0) // width]
        text(f'Welcome on board with {welcome_msg[airline_key]}! You just received 1 point!', 80, 130, 20)
        Score += 1
        background[(yh - y0) // width][(xh - x0) // width] = '_'
        
    elif background[(yh - y0) // width][(xh - x0) // width] == 'M':
        clear(200, 200, 200)
        text('Game Over', 170, 200, 80)
        text('Score: {}'.format(Score), 280, 300, 50)
        wait_mouse_leftclick()
        all_done()
    elif background[(yh - y0) // width][(xh - x0) // width] == 'B':
        background[(yh - y0) // width][(xh - x0) // width] = '_'
        clear(200, 200, 200)
        text('Welcome on board with British Airways! You just received 1 point!', 80, 130, 20)
        draw('MazeGame_imgs/I.png',10,200)
        #The player needs to help British Airways to make a decision through pushing 'h' or 'i'. 
        text('However, now British Airways has an employee work crisis. ',10,480,20)
        text('If you help them to make the decision, you can get extra points! ',10,500,20)
        text('Please make your decision by inputing "h" or "i"',10,560,20)
        text('    h. Hire new staff ',10,520,20)
        text('    i. Increase existing staff pay',10,540,20)
        while True:
            choice=wait_key_press()
            if choice =='h':
                draw_background()
                human(xh,yh)
                text('Thank you for supporting hiring new staffs! You recived extre 5 points!', 80, 170, 20)
                Score += 5
                break
            elif choice =='i':
                draw_background()
                human(xh,yh)
                text('Thank you for increasing existing staff pay! You recived extre 5 points!', 80, 170, 20)
                Score += 5
                break
                
                
            ##Design two different rewards (no more than 5 points) to the player in response to the different decision. 

    
            
game_running = True
point=0
xh=40
yh=230
    
while game_running:
    draw_background()
    human(xh,yh)
    while True:
        userinput = wait_key_press()
        if userinput == "a":
            if xh<70:   #limit the border
                continue
            else:
                draw_background()
                xh-=30
                human(xh,yh)
            text('Score: {}'.format(Score), 10, 50, 20)
              
        elif userinput == "d":
            if xh>470:
                continue
            else:
                draw_background()
                xh+=30
                human(xh,yh)
            text('Score: {}'.format(Score), 10, 50, 20)
                
        elif userinput == "w":
            if yh<260:
                continue
            else:
                draw_background()  
                yh-=30
                human(xh,yh)
            text('Score: {}'.format(Score), 10, 50, 20)
               
        elif userinput == "s":
            if yh>260:
                continue
            else:
                draw_background()
                yh+=30
                human(xh,yh)
            text('Score: {}'.format(Score), 10, 50, 20)
                
        elif userinput == "q":
            draw_background()
            text('Game Over',300,130,20)
            game_running = False
            break
            wait_mouse_leftclick()
            all_done()
        else: 
            text('Invalid Input', 80, 130, 20)
        
        if Score>=20:
            clear(200, 200, 200)
            draw('MazeGame_imgs/you_win.png',130,120)
            text('You win!', 250, 20, 70)
            text('Score: {}'.format(Score), 320, 100, 30)
            wait_mouse_leftclick()
            all_done()
        
    



    

wait_mouse_leftclick()
all_done()
```
