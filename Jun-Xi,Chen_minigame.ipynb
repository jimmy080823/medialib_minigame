```python
def image_file_path(filename):
      return f"MazeGame_imgs/{filename}.png"

def print_map(map):
      x = 10
      y = 200
      offset = 30
      
      for row in map:
            x = 10
            for cell in row:
                  draw(image_file_path(cell), x, y)
                  x += offset
            y += offset 
```

```python
from medialib import *
initialize() # always the first instruction of the program
##################################################################################


##  each item of the list representing the background of each row
background=[['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X'],
            ['X','_','_','L','_','_','M','_','_','E','_','_','J','_','_','_','_','X'],
            ['X','_','_','_','M','_','B','_','L','_','_','_','_','B','_','J','_','X'],
            ['X','_','_','B','_','_','_','_','_','J','_','_','E','_','L','_','_','X'],
            ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X']] 

airplane = {
    'B': 'British Airways',
    'E': 'EasyJet',
    'L': 'Lufthansa',
    'J': 'Jet2'
}

clear(200,200,200) ## background
print_map(background)
text('Maze game',300,100,20)
text('You are a lost hero, and have to find your way back home...',10,500,16)
text('When you collect 20 points, you can get back to home!',10,520,16)
text('To play: press a and d keys to move left and right',10,540,16)
text('         press w and s keys to move upwards and downwards',10,560,16)
text('         press q to quit the game',10,580,16)
    


collidables = 'X' ## the letters that the avatar cannot pass
monsters= 'M' ## the letters that the avatar will dead when meeting it
point = 0
## game state
game_running = True
while game_running:
    ##The avatar will dead when meeting a monster represent by ‘M’. 
    
    ## If the avatar stand on a position which originally has Lufthansa/easyJet/Jet2 airplane (L.png/E.png/J.png), 
    # then the flight will be taken by the avatar.
    # A welcome message 'Welcome on board with ??? ! You just received 1 point!' 
    ## When the avatar leave that position,the image of airplane logo will be changed to the image of road(_.png).
    
    
    ## If the avatar stand on a position which originally has British Airways(B.png), 
    # then the flight will be taken by the avatar.
    # A welcome message 'Welcome on board with British Airways ! You just received 1 point!' 
    
    ## When the avatar leave that position,the image of the airplane(B.png) will be changed to the image of road(_.png).
    

    
        #The player needs to help British Airways to make a decision through pushing 'h' or 'i'. 
        # draw('MazeGame_imgs/I.png', 10,200)
        # text('However, now British Airways has an employee work crisis. ',10,480,20)
        # text('If you help them to make the decision, you can get extra points! ',10,500,20)
        # text('Please make your decision by inputing "h" or "i"',10,560,20)
        # text('    h. Hire new staff ',10,520,20)
        # texts('    i. Increase existing staff pay',10,540,20)
    
    #player init position
    cur_x = 1
    cur_y = 1
    msg = ''
    while True:
        #display
        text(f'Current Point: {point}', 0, 0, 16)
        text(msg,50,100,20)
        background[cur_x][cur_y] = 'H'
        msg = ''
        print_map(background)

        #choice = wait_key_press()
        ##Design two different rewards (no more than 5 points) to the player in response to the different decision. 
  
        ## check and react to input
        key = wait_key_press()



        ## if the player push 'q', quit the game. 
        if key.lower() == 'q':
            clear(200, 200, 200)
            text('Game Over',230, 250, 50)
            text('Your Point: {}'.format(point), 210, 330, 50)
            game_running = False
            break

        ## if the player push 'a',the avatar go left.
        ## update game state
        elif key.lower() == 'a':
            new_x = cur_x
            new_y = cur_y - 1
    
        ## if the player push 'd',the avatar go right.
        ## update game state
        elif key.lower() == 'd':
            new_x = cur_x
            new_y = cur_y + 1
            
        ## if the player push 'w',the avatar go upward.
        ## update game state
        elif key.lower() == 'w':
            new_x = cur_x - 1
            new_y = cur_y
            
        ## if the player push 's',the avatar go downward.
        ## update game state
        elif key.lower() == 's':
            new_x = cur_x + 1
            new_y = cur_y

        #validate path
        cell = background[new_x][new_y]

        if cell == 'X':
            continue
        elif cell == 'E' or cell == 'J' or cell == 'L':
            msg = f'Welcome on board with {airplane[cell]}! You just receive 1 point!'
            point += 1
        elif cell == 'M':
            clear(200, 200, 200)
            text('You Die!  Game Over!!',110, 250, 50)
            text('Your Point: {}'.format(point), 210, 330, 50)
            game_running = False
            break
        elif cell == 'B':
            
            clear(200,200,200)
            text('Welcome on board with British Airways! You just received 1 point!', 80, 130, 20)
            draw('MazeGame_imgs/I.png',10,200)
            #The player needs to help British Airways to make a decision through pushing 'h' or 'i'. 
            text('However, now British Airways has an employee work crisis. ',10,480,20)
            text('If you help them to make the decision, you can get extra points! ',10,500,20)
            text('Please make your decision by inputing "h" or "i"',10,560,20)
            text('    h. Hire new staff ',10,520,20)
            text('    i. Increase existing staff pay',10,540,20)
            
            while True:
                choice=wait_key_press()
                if choice =='h':
                    print_map(background)
                    text('Thank you for supporting hiring new staffs! You recived extre 5 points!', 80, 170, 20)
                    point += 5
                    break
                elif choice =='i':
                    print_map(background)
                    text('Thank you for increasing existing staff pay! You recived extre 5 points!', 80, 170, 20)
                    point += 5
                    break

        if point>=20:
            clear(200, 200, 200)
            draw('MazeGame_imgs/you_win.png',130,120)
            text('You win!', 250, 20, 70)
            text('Point: {}'.format(point), 320, 100, 30)
            wait_mouse_leftclick()
            all_done()
               
            
        #update background
        background[cur_x][cur_y] = '_'
        cur_x = new_x
        cur_y = new_y
        clear(200,200,200)


wait_mouse_leftclick()
all_done()
```
